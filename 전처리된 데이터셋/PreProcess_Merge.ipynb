```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='median')
store['CompetitionDistance'] = imputer.fit_transform(store[['CompetitionDistance']])
```

```python
store.fillna(0, inplace=True)
store.isnull().sum()
```

```python
train_store = pd.merge(train, store, how='left', on='Store')
train_store.info()
```

```python
# 'Date' 컬럼을 datetime으로 변환
train_store['Date'] = pd.to_datetime(train_store['Date'])

# Date 컬럼 위치 확인 후 Year, Month, Day 컬럼 추가
date_col_index = train_store.columns.get_loc('Date')
train_store.insert(date_col_index + 1, 'Year', train_store['Date'].dt.year)
train_store.insert(date_col_index + 2, 'Month', train_store['Date'].dt.month)
train_store.insert(date_col_index + 3, 'Day', train_store['Date'].dt.day)

# 결과 확인
train_store[['Date', 'Year', 'Month', 'Day']].head()
```
